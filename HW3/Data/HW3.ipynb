```python
#Import Libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load in linear separable text. 
x_linsep = np.loadtxt('X_LinearSeparable.txt', delimiter = ',')
y_linsep = np.loadtxt('Y_LinearSeparable.txt', delimiter = ',')
```

```python
#Note: x_linsep has 3 samples of length 20 arrays. Need to change this. 
x_linsep = np.transpose(x_linsep)
```

```python
#1a. Perception Learning Algorithm
w = np.zeros(len(x_linsep[0])) #Number of training samples.
for i in range(100): #100 iterations. 
    for i, j in enumerate(x_linsep): #PLA Algorithm. 
        if(np.dot(w, x_linsep[i]) * y_linsep[i]) <= 0: #w^Tx = Dot product. Multiply with y_linsep to get sign. 
            w = w + x_linsep[i]*y_linsep[i] 
print(w) #Weight vector
```

```python
#1b. Plotting
for i, j in enumerate(x_linsep):
    if(y_linsep[i] == 1):
        plt.scatter(x_linsep[i][1], x_linsep[i][2], c='orange') #y = 1. No need to include first element as it is all 1. 
    else:
        plt.scatter(x_linsep[i][1], x_linsep[i][2], c='blue') #y = 0
#Rotate w vector 90 degrees about the origin to get decision boundary. (x, y) -> (-y, x)
#Alternatively, compute -w[0]/w[2] and -w[1]/w[2] to get y-intercept and slope respectively. They should be identical for now. 
x = [0, -w[2]]
y = [-w[0]/w[2], w[1]]
plt.xlim(0, 1.1)
plt.ylim(0, 1.1)
plt.plot(x, y, c='black')
```

```python
#1c. Error Checking.
N_mistakes = 0
for i, j in enumerate(x_linsep):
    if(np.dot(w, x_linsep[i]) < 0 and y_linsep[i] != -1): #If sign(w^Tx = -1 and y_linsep does not equal -1)
        N_mistakes += 1
    elif(np.dot(w, x_linsep[i]) > 0 and y_linsep[i] != 1): #If sign(w^Tx = 1 and y_linsep does not equal 1)
        N_mistakes += 1
print('Number of Mistakes: ' + str(N_mistakes))
print('Error Rate: ' + str(N_mistakes / y_linsep.size)) #N_mistakes divided by number of elements.  
```

```python
#Load in handwritten digits training. 
x_hand_digits_train = np.loadtxt('X_Digits_HandcraftedFeature_Train.txt', delimiter = ',')
y_hand_digits_train = np.loadtxt('Y_Digits_HandcraftedFeature_Train.txt', delimiter = ',')
```

```python
#Note: x_hand_digits_train has 3 samples of length 1561 arrays. Need to change this. 
x_hand_digits_train = np.transpose(x_hand_digits_train)
```

```python
#3a. Perception Learning Algorithm
w_pocket = np.zeros(len(x_hand_digits_train[0])) #Number of training samples.
for i in range(100): #100 iterations. 
    for i, j in enumerate(x_hand_digits_train): #PLA Algorithm. 
        if(np.dot(w_pocket, x_hand_digits_train[i]) * y_hand_digits_train[i]) <= 0: #w^Tx = Dot product. Multiply with y_hand_digits to get sign. 
            w_pocket += x_hand_digits_train[i]*y_hand_digits_train[i] 
print(w_pocket) #Weight vector
```

```python
#3b. Testing
#Load in handwritten digits testing.
x_hand_digits_test = np.loadtxt('X_Digits_HandcraftedFeature_Test.txt', delimiter = ',')
y_hand_digits_test = np.loadtxt('Y_Digits_HandcraftedFeature_Test.txt', delimiter = ',')
```

```python
#Note: x_hand_digits_test has 3 samples. Need to change this. 
x_hand_digits_test = np.transpose(x_hand_digits_test)
```

```python
#Create Y_test_pred
Y_test_pred = np.zeros(y_hand_digits_test.size) #Initialize array with just zeroes.
for i, j in enumerate(x_hand_digits_test):
    if(np.dot(w_pocket, x_hand_digits_test[i]) > 0): 
        Y_test_pred[i] = 1 #Sign Function = 1
    elif(np.dot(w_pocket, x_hand_digits_test[i]) < 0):
        Y_test_pred[i] = -1 #Sign Function = -1
    else:
        Y_test_pred[i] = 0
```

```python
#Compare with Y_test
N_mistakes = 0
for i, j in enumerate(y_hand_digits_test):
    if(Y_test_pred[i] != y_hand_digits_test[i]):
        N_mistakes += 1
print('Number of Mistakes: ' + str(N_mistakes))
print('Error Rate: ' + str(N_mistakes / y_hand_digits_test.size)) #N_mistakes divided by number of elements. 
```

```python
#3c. Plotting
for i, j in enumerate(x_hand_digits_test):
    if(y_hand_digits_test[i] == 1):
        plt.scatter(x_hand_digits_test[i][1], x_hand_digits_test[i][2], c='orange') #y = 1. No need to include first element as it is all 1. 
    else:
        plt.scatter(x_hand_digits_test[i][1], x_hand_digits_test[i][2], c='blue') #y = 0
#Rotating 90 degrees using geometry no longer works. You have to use intercept/slope formula. 
intercept = -w_pocket[0]/w_pocket[2]
slope = -w_pocket[1]/w_pocket[2]
x = [0, 1]
y = [intercept, intercept + slope]
plt.xlim(0, 1)
plt.ylim(-8, 0)
plt.plot(x, y, c='black')
```
